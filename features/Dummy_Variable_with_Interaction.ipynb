```python
import pandas as pd
cat_data = pd.read_excel('dummy_variable.xlsx')
cat_data.head()
```

```python
df = pd.get_dummies(cat_data, columns=['Education', 'Managerial_Pos'],drop_first=True)
df.head()
```

```python
import statsmodels.api as sm
X = df[['Exp','Education_PhD','Education_Post-Graduate','Managerial_Pos_Yes']] 
Y = df['Salary']
X = sm.add_constant(X) 

model = sm.OLS(Y, X).fit()
print_model = model.summary()
print(print_model)
```

```python
# interaction_df = pd.DataFrame()
df['Phd_Mgr'] = df['Education_PhD'] * df['Managerial_Pos_Yes']
df['PG_Mgr'] = df['Education_Post-Graduate'] * df['Managerial_Pos_Yes']
df.head()
```

```python
X = df[['Exp','Education_PhD','Education_Post-Graduate','Managerial_Pos_Yes','Phd_Mgr','PG_Mgr']] 
Y = df['Salary']
X = sm.add_constant(X) 

model = sm.OLS(Y, X).fit()
print_model = model.summary()
print(print_model)
```
